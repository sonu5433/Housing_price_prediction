# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
err=cross_val_score(linreg,x_train,y_train,cv=3)
err=-err
print(f"mean: {err.mean()} \n std : {err.std()}")
print("Falat")
err=cross_val_score(linreg,x_train,y_train2,cv=3)
err=-err
print(f"mean: {err.mean()} \n std : {err.std()}")


```

 # Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
err=cross_val_score(model,x_train,y_train,cv=3)
err=-err
print(f"mean: {err.mean()} \n std : {err.std()}")
print("Falat")
err=cross_val_score(model,x_train,y_train2,cv=3)
err=-err
print(f"mean: {err.mean()} \n std : {err.std()}")



```

# RandomForest

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=10)
err=cross_val_score(model,x_train,y_train,cv=3)
err=-err
print(f"mean: {err.mean()} \n std : {err.std()}")
print("Falat")
err=cross_val_score(model,x_train,y_train2,cv=3)
err=-err
print(f"mean: {err.mean()} \n std : {err.std()}")
```
